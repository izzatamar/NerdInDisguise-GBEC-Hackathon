```python
import datetime
import hashlib
from flask import Flask, request
from flask import jsonify
import json

class Block:
    blockNo = 0
    data = None
    next = None
    hash = None
    nonce = 0
    previous_hash = 0x0
    timestamp = None

    def __init__(self, data):
        self.data = data

    def hash(self):
        self.timestamp = datetime.datetime.now()
        h = hashlib.sha256()
        h.update(
        str(self.nonce).encode('utf-8') +
        str(self.data).encode('utf-8') +
        str(self.previous_hash).encode('utf-8') +
        str(self.timestamp).encode('utf-8') +
        str(self.blockNo).encode('utf-8')
        )
        return h.hexdigest()

    def __str__(self):
        blockChainDesc = "Block Hash: " + str(self.hash()) + "\nBlockNo: " + str(self.blockNo) + "\nBlock Data: " + str(self.data) + "\nHashes: " + str(self.nonce) + "\n--------------"
        return blockChainDesc

class Blockchain:

    diff = 20
    maxNonce = 2**32
    target = 2 ** (256-diff)

    block = Block("Genesis")
    dummy = head = block

    def add(self, block):

        block.previous_hash = self.block.hash()
        block.blockNo = self.block.blockNo + 1

        self.block.next = block
        self.block = self.block.next

    def mine(self, block):
        for n in range(self.maxNonce):
            if int(block.hash(), 16) <= self.target:
                self.add(block)
                print(block)
                break
            else:
                block.nonce += 1



blockchain = Blockchain()

app = Flask(__name__)

dictList = []

@app.route('/data',methods=['GET'])
def json_output():

    return jsonify(dictList)



@app.route('/detection', methods=['POST']) #GET requests will be blocked
def json_example():
    dataDict ={"publicKey": None,
           "age":None,
           "gender": None,
           "location": None,
           "angry":None,
           "disgust": None,
           "fear" :None,
           "happy": None,
           "sad": None,
           "surprise": None,
           "neutral":None,
              "hash":None} 

    req_data = request.get_json()
    dataDict["publicKey"] = req_data['publicKey'] 
    dataDict["age"] = req_data['age']
    dataDict["gender"] = req_data['gender']
    dataDict["location"] = req_data['location']
    #emotion = req_data['emotion'] 
    dataDict["angry"] = req_data['angry'] 
    dataDict["disgust"] = req_data['disgust'] 
    dataDict["fear"] = req_data['fear'] 
    dataDict["happy"] = req_data['happy'] 
    dataDict["sad"] = req_data['sad'] 
    dataDict["surprise"] = req_data['surprise'] 
    dataDict["neutral"] = req_data['neutral'] 
    dataDict["hash"] = blockchain.block.hash()
    dataDict["previous_hash"] = blockchain.block.previous_hash
    dataDict['time'] = blockchain.block.timestamp
    blockchain.mine(Block(dataDict)) 
    
    dictList.append(dataDict) 
  
if __name__ == "__main__":
    app.run(host='192.168.43.5')
    
while blockchain.head != None:
    print("\ndata: ",blockchain.head.data)
    dataList.append(blockchain.head.data)
    #dataDict["hash"] = blockchain.head.hash()
    blockchain.head=blockchain.head.next
     

```
